```python
# Importing required libraries
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Loading the dataset
df=pd.read_csv('Diwali.csv',encoding='unicode_escape')
df
```

```python
df.shape
```

```python
df.info()
```

```python
# Drop null columns
df.drop(['Status', 'unnamed1'],axis=1 , inplace=True)
```

```python
df.info()
```

```python
# check for null values
pd.isnull(df).sum()
```

```python
# drop null values
df.dropna(inplace=True)
```

```python
# change data type
df['Amount']=df['Amount'].astype('int')
```

```python
df['Amount'].dtypes
```

```python
df.describe()
```

```python
# Exploratory Data Analysis
df.columns
```

```python
# To get number of male and female shopper

x = sns.countplot(x='Gender', data=df)
x.bar_label(x.containers[0])

```

```python
# Total sales by each gender
sales_gen = df.groupby('Gender', as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
sales_gen

```

```python
sns.barplot(x='Gender',y='Amount',data=sales_gen)
```

```python
# To find shoppers of diffrent age groups in both the genders
df.columns
```

```python
Age=sns.countplot(data=df,x='Age Group',hue="Gender")
```

```python
# Total sales by each age group
sales_gen = df.groupby('Age Group', as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
sales_gen
```

```python
sns.barplot(x='Age Group',y='Amount',data=sales_gen)
```

```python
# top 10 states from where the highest number of orders were placed
df.columns
```

```python
states = df.groupby('State', as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)
states
```

```python
sns.barplot(x='State',y='Orders',data=states)
```

```python
# top 10 states from where the sales were highest 
df.columns
```

```python
states = df.groupby('State', as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)
states
```

```python
sns.barplot(x='State',y='Amount',data=states)

sns.set(rc={'figure.figsize':(20,9)})
```

```python
# Shoppers with Different relationship status
df.columns
```

```python
# To get relationship status of shoppers

x = sns.countplot(x='Marital_Status', data=df)
x.bar_label(x.containers[0])
```

```python
sales_RS = df.groupby('Marital_Status', as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)
sales_RS
```

```python
sns.barplot(x='Marital_Status',y='Amount',data=sales_RS)

plt.figure(figsize=(15,5))
```
